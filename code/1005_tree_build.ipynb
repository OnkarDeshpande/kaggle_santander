```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

import graphviz
# sudo apt-get install graphviz
```

```python
train_file = '../data/train.csv'
df_train = pd.read_csv(train_file)

target = df_train.pop('target')
df_train.drop(['ID_code'], axis=1, inplace=True)
```

```python
train_X, val_X, train_y, val_y = train_test_split(df_train, target, random_state=1)
tree_model = DecisionTreeClassifier(random_state=0, max_depth=5, min_samples_split=5).fit(train_X, train_y)

tree_graph = tree.export_graphviz(tree_model, out_file=None, feature_names=list(df_train.columns))
```

```python
graphviz.Source(tree_graph)
```
