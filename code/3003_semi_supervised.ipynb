```python
import numpy as np
import pandas as pd

import seaborn as sns
sns.set_style('whitegrid')

import time

import lightgbm as lgb

from sklearn.model_selection import KFold,StratifiedKFold
from sklearn.model_selection import GridSearchCV

from sklearn.semi_supervised import LabelPropagation, LabelSpreading
```

```python
train_file = '../data/train.csv'
test_file = '../data/test.csv'

df_train = pd.read_csv(train_file)
df_test = pd.read_csv(test_file)

np.random.seed(42)
```

```python
##create data
df_train.drop(['ID_code'], axis=1, inplace=True)
df_test.drop(['ID_code'], axis=1, inplace=True)

df_test['target'] = -1
df_data = pd.concat([df_train[:150000], df_test[:50000]], sort=False)

df_data.reset_index(drop=True, inplace=True)
```

```python
y=df_data.pop('target')
```

```python
df_data.tail()
```

```python
lp_rbf = LabelPropagation(kernel='rbf', gamma=9e-6, max_iter=10000, tol=0.0001)
lp_rbf.fit(df_data, y)
```
